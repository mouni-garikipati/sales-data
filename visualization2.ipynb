```python
def remove_group_outliers(df, group_col, target_col):
    df_clean = pd.DataFrame()
    for name, group in df.groupby(group_col):
        Q1 = group[target_col].quantile(0.25)
        Q3 = group[target_col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        group_filtered = group[(group[target_col] >= lower) & (group[target_col] <= upper)]
        df_clean = pd.concat([df_clean, group_filtered])
    return df_clean

# Apply function
df_clean_group = remove_group_outliers(df_no_outliers, 'TERRITORY', 'SALES')
```

```python
sns.boxplot(data=df_clean_group, x='TERRITORY', y='SALES')
plt.title('Sales by Territory (Outliers Removed per Group)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
